```python
from quartic_sdk import APIClient
from quartic_sdk.core.entity_helpers.entity_list import EntityList
import os

client = APIClient(os.environ.get('QUARTIC_HOST'), username=os.environ.get('QUARTIC_USERNAME'), password=os.environ.get('QUARTIC_PASSWORD'))
```

### Asset

```python
#Entity List of all available assests
assets=client.assets()
print("First 5 asssets from EntityList of available assets : \n" , assets[0:5])

#EntityList of required assets
asset1=assets.get("id",865)
asset2=assets.get("id",864)
asset_list = EntityList("Asset")
asset_list.add(asset1)
asset_list.add(asset2)
print("\nasset_list : ", asset_list)

#Get data of all tags of an asset between specified duration
asset1_data = asset1.data(start_time=1617086964262,stop_time=1617096964262)
print("\nasset1_data : ")
asset1_data[0] #Pandas df
```

### Tag

```python
#EntityList of all tags of a particular Asset
tags = asset1.get_tags()
print("First 5 tags from EntityList of all available tags of asset1 : \n", tags[0:5])

#Get data of all tags in entitylist of tags between specified duration
tags_data = tags.data(start_time=1617086964262,stop_time=16170969642621)
print("\nTag data of all tags : ")
tags_data[0] #pandas df
```

```python
tag1=tags.get("id",39938)
tag2=tags.get("id",39939)

#EntityList of required tags
tag_list=EntityList("Tag")
tag_list.add(tag1)
tag_list.add(tag2)
tag_list_data = tag_list.data(start_time=1617086964262,stop_time=16170969642621)
print("Tag data of selected tags : ")
tag_list_data[0] #pandas df
```

### Batch

```python
#EntityList of all batches of an asset
batches = asset1.batches()
print("First 5 batches from EntityList of all available batches of an asset : \n", batches[0:5])
print()

batch_at_index_0 = batches[0]
batch_at_index_1 = batches[1]

#EntityList of required batches
batch_list = EntityList("Batch")
batch_list.add(batch_at_index_0)
batch_list.add(batch_at_index_1)
print("\nbatch_list : ", batch_list)
```

### Edge Connectors

```python
#EntityList of all available edge_connectors
edge_connectors = client.edge_connectors()
print("All available edge connectors : \n", edge_connectors)
edge_connector_at_index_0 = edge_connectors[0]
edge_connector_at_index_1 = edge_connectors[1]


#EntityList of all tags of a specific edge_connector
ec_at_index_0_tags = edge_connector_at_index_0.get_tags()
print("\nFirst 5 tags from EntityList of all tags of an edge connector : \n", ec_at_index_0_tags[0:5])

#EntityList of required edge_connectors
ec_list = EntityList("EdgeConnector")
ec_list.add(edge_connector_at_index_0)
ec_list.add(edge_connector_at_index_1)
print("\nec_list : \n", ec_list)

#Get data of all tags of an edge_connector
ec_data = edge_connector_at_index_0.data(start_time=1617086964262,stop_time=161709696426)
print("\nec_data : ")
ec_data[0]
```

### Cotext Frames and ContextFrame Occurances

```python
#EntityList of all available Context Frames
context_frames = client.context_frames()
print("EntityList of all available context Frames : \n", context_frames)

#EntityList of all occurences of a particular ContextFrame
cf_at_index_0 = context_frames[0]
cfo = cf_at_index_0.occurrences()
print("\nFirst 5 occurances from EntityList of all occurences of a ContextFrame : \n", cfo[0:5])
```

```python
cfo_at_index_0, cfo_at_index_1, cfo_at_index_2 = cfo[0], cfo[1], cfo[2]
#EntityList of specific ContextFrame Occurances
cfo_list = EntityList("ContextFrameOccurrence")
cfo_list.add(cfo_at_index_0)
cfo_list.add(cfo_at_index_1)
cfo_list.add(cfo_at_index_2)
print("EntityList of selected ContextFrame Occurances : \n", cfo_list)
```
